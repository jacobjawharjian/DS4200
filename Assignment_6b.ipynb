```python
import pandas as pd
import altair as alt

listed_species_df = pd.read_csv("species-listings-by-year-totals-report.csv")
listed_species_df.head()
```

```python
delisted_species_df = pd.read_csv("FWS_Species_Data_Explorer (1).csv")

delisted_species_df['Delisting Date'] = delisted_species_df['Delisting Date'].str[6:]
delisted_species_df = delisted_species_df.drop(['Scientific Name', 'Scientific Name_url', 'Entity Description', 'ESA Listing Date', 'Delisting Reason', 'Foreign or Domestic', 'Lead Agency'], axis=1)

remove_count_df = delisted_species_df.groupby(['Delisting Date'], as_index=False).count()
remove_count_df.head()
```

## Excellent

```python
listed = alt.Chart(listed_species_df, title="Number of Species Added to the Endangered/Threatened List").mark_line(point=True).encode(
    x=alt.X("Calendar Year:O", title="Year").scale(zero=False),
    y=alt.Y("Number of Species Listings:Q", title="Species Added").scale(zero=False), 
    tooltip = ['Number of Species Listings']
)

listed
```

## Deceptive

```python
listed = alt.Chart(listed_species_df, title="Number of Species Added to the Endangered/Threatened List").mark_line(point=False, color='green').encode(
    x=alt.X("Calendar Year:O", title="Year").scale(zero=False),
    y=alt.Y("Number of Species Listings:Q", title="Species Added").scale(zero=False), 
).properties(
    width=800
    
)

delisted = alt.Chart(remove_count_df, title="Number of Species Removed from the Endangered/Threatened List").mark_line(point=False, color='red').encode(
    x=alt.X("Delisting Date:O", title="Year").scale(zero=False),
    y=alt.Y("Common Name:Q", title="Species Removed").scale(zero=False)
).properties(
    width=800  
)


listed & delisted
```
