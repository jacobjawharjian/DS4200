## Custom Ticks

```python
#Quick way of creating a figure and axes:
fig, ax = plt.subplots()

#Locator handles the position of ticks and tick labels
#Formatter handles the customization or formatting of tick labels

#Import locators from matplotlib.ticker
from matplotlib.ticker import AutoLocator, MaxNLocator, LinearLocator, MultipleLocator, FixedLocator, NullLocator

ax.set_xlim(0,10)
ax.set_ylim(0,5)

#Create a handle for x and y axis
x=ax.xaxis
y=ax.yaxis
```

```python
max_l = MaxNLocator(4)
max_t = MaxNLocator(10)

#Pass to the axes
x.set_major_locator(max_l)
y.set_major_locator(max_t)
fig
```

```python
lin_loc = LinearLocator(11)

#Pass it to the x and y axis
x.set_major_locator(lin_loc)
fig
```

```python
mul_loc = MultipleLocator(4)

#Pass it to the x and y axis
x.set_major_locator(mul_loc)
y.set_major_locator(mul_loc)
fig
```

## Customizations

```python
fig2, axes2 = plt.subplots(figsize=(10,4)) 

#Add color, alpha(transparency),linewidth,linestyle, marker, marker size, marker color
axes2.plot(x1,y1,'#aa556f', alpha=0.5, lw=2, ls='--', marker='x', markersize='10', 
           markeredgecolor='r', markeredgewidth=4)

#Set lower and upper bounds/ Zooming in: 
axes2.set_xlim([0,8])
axes2.set_ylim([0,6])

#Customize the ticks 
axes2.tick_params(axis='x', labelcolor='blue', labelsize=20)
axes2.tick_params(axis='y', labelcolor='blue', labelsize=20)

plt.show()
```

## Pandas

```python
#Upload data using Pandas library (csv)
df = pd.read_csv('iris.csv')
df.head() #Will return first 5 rows of the dataset
#df #Will return the whole dataset
```

```python
#To get some descriptive statistics about the dataset:  
df.describe()

#Display certain rows only
df[0:4]

#Display certain columns only
df[['petal_width', 'petal_length']]


#Display certain rows and columns only (iloc[rows, columns])
df.iloc[0:4, 0:2]

#Similar to iloc, use loc to either use labels or numeric representation
df.loc[0:4, ['petal_width', 'petal_length']]

df.loc[df["sepal_width"] == 3.0]

#Want to change the column names:
df.columns = 'Sepal length', 'Sepal width', 'Petal length', 'Petal width', 'Species'
df

#Modify the data by adding columns (can do other calculations as well)
df["Sepal area"] = df["Sepal length"] * df["Sepal width"]
df["flower?"] = "yes"
df.head()

#Delete data
df=df.drop(columns=['flower?'])
df
```

```python
df["Id"] = list(range(len(df)))
df
```

```python
#Instantiate figure and axis
fig, ax = plt.subplots(figsize=(10, 4))

#If you don't pass any x-axis values then by default the index will be the x-axis values
ax.plot('Sepal length', data=df, label='length', color='c', lw=2)

#add another line chart
ax.plot('Sepal width', data=df, label='width', color='r', lw=2)

#Add legend
ax.legend()

#Add title and axes labels
ax.set_title('Sepal dimensions', size='20')
ax.set_ylabel('measurement in cm', size='15')
ax.set_xlabel('index', size='15')

#Customize ticks and labels
from matplotlib.ticker import StrMethodFormatter
ax.tick_params(axis='both', labelsize='10', labelcolor='k')

y_lab = StrMethodFormatter("{x:.1f} cm")
ax.yaxis.set_major_formatter(y_lab)

#Add markers to the first line
line1= ax.lines[0]

line1.set_marker('o')
line1.set_markevery(10)
line1.set_markerfacecolor('white')
line1.set_markeredgewidth(1)
line1.set_markeredgecolor('c')
#shortcut:
#line1.set(marker='o', markevery=10, markerfacecolor='white')

#Add markers to second line
line2= ax.lines[1]
line2.set(marker='o', markevery=10, markerfacecolor='white', markeredgewidth=1, markeredgecolor='r')

#Find min and max length 
val = df['Sepal length'].values
l_min = val.min()
l_max = val.max()
#Find the corresponding arguments in the x-axis
x_min = val.argmin()
x_max = val.argmax()

#Draw vertical lines to indicate min and max
ax.axvline(x_min, color='c', ls='--', lw=2)
ax.axvline(x_max, color='c', ls='--', lw=2)

#Find min and max width 
val = df['Sepal width'].values
w_min = val.min()
w_max = val.max()
#Find the corresponding arguments in the x-axis
x_min = val.argmin()
x_max = val.argmax()

#Draw vertical lines to indicate min and max
ax.axvline(x_min, color='r', ls='--', lw=2)
ax.axvline(x_max, color='r', ls='--', lw=2)

plt.show()
```

```python
df_setosa = df.loc[df["Species"] == "setosa"]
df_versicolor = df.loc[df["Species"] == "versicolor"]


df["Sample"] = df["Id"] % 50
df.iloc[47:55]

df_versicolor.head()
```

```python
fig3, ax3 = plt.subplots(figsize=(10,6)) 
ax3.bar('Sample', 'Petal length', data=df_setosa, label='setosa', color='#71a376')
ax3.bar('Sample', 'Petal length', data=df_versicolor, label='versicolor', color='#cfd8b2', zorder=-1)
# ax3.bar('Sample', 'Petal length', data=df_versicolor, label='versicolor', color='#cfd8b2', bottom=df_setosa["Petal length"])

ax3.set_title('Petal length of two flower species', size=20)
ax3.set_ylabel('Petal length (cm)')
ax3.set_xlabel('Sample No.')
fig3.legend()
ax3.set_xlim(-.5,49.5)
plt.show()
```

```python
fig4, ax4 = plt.subplots(figsize=(14,6))

x_vals = np.arange(len(df_setosa["Sample"]))
width = 0.35

ax4.bar(x_vals - width / 2, 'Petal length', data=df_setosa, label='setosa', color='#71a376', width=width)
ax4.bar(x_vals + width / 2, 'Petal length', data=df_versicolor, label='versicolor', color='#cfd8b2', width=width)
ax4.set_xticks(x_vals, df_setosa.Sample)

ax4.set_title('Petal length of two flower species', size=20)
ax4.set_ylabel('Petal length (cm)')
ax4.set_xlabel('Sample No.')
fig4.legend()
ax4.set_xlim(-1,50)
plt.show()
```
