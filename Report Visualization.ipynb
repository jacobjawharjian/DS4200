```python
import pandas as pd
import altair as alt

census_df = pd.read_csv('CleanedCharlestownCensusData_WithFamilyCounts.csv')

census_df['Occupation'].value_counts().head(12)
```

```python
census_df = census_df.loc[(census_df["Occupation"] == "Sea captain") | (census_df["Occupation"] == "Baker") | (census_df["Occupation"] == "Leather dresser") | (census_df["Occupation"] == "Cooper") | (census_df["Occupation"] == "Blacksmith") | (census_df["Occupation"] == "Housewright") | (census_df["Occupation"] == "Hatter") | (census_df["Occupation"] == "Chaise maker") | (census_df["Occupation"] == "Mariner") | (census_df["Occupation"] == "Cordwainer")]
census_df.head()
```

```python
occupations = census_df['Occupation'].unique()
colors = ['#77AADD', '#99DDFF', '#44BB99', '#BBCC33', '#AAAA00', '#EEDD88', '#EE8866', '#FFAABB', '#555555', '#CC6677']
occupations
```

```python
census_chart = alt.Chart(census_df, title="Distribution of Family Size per Worker in an Occupation (10 most common occupations)").mark_boxplot(opacity=0.75).encode(
    x=alt.X("Occupation"),
    y="Number of Family Members (Including Head):Q",
    color=alt.Color('Occupation:N', scale=alt.Scale(domain=occupations, range=colors))
    
).properties(
    width=450,
    height=250
)



avg_chart = alt.Chart(census_df, title="Average Family Size per Worker in an Occupation (10 most common occupations)").mark_bar(opacity=0.75, color='red').encode(
    x=alt.X("Occupation:N"),
    y="average(Number of Family Members (Including Head))",
    color=alt.Color('Occupation:N', scale=alt.Scale(domain=occupations, range=colors))
    
).properties(
    width=450,
    height=250
)



census_chart & avg_chart
```
